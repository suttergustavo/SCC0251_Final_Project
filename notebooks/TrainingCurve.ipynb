```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
train_size = 5000
epochs = 20
```

```python
loss_history = np.load('../loss_history.npy')
iterations = np.linspace(0, train_size * epochs, loss_history.shape[0])
```

```python
plt.figure(figsize=(15, 10))
plt.plot(iterations, loss_history)
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Loss (log scale)')
plt.title('Training loss history')
plt.show()
```
